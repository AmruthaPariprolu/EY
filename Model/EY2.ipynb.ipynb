```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Assuming df_cleaned is your dataset with relevant columns ('Skills', 'URL', 'Category', 'Sub-Category')

# Create a TfidfVectorizer to convert the 'Skills' text into numerical vectors
vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the Skills column
skills_matrix = vectorizer.fit_transform(df['Skills'])

# Function to get recommendations based on user input
def recommend_courses(user_input):
    # Transform the user input into the same vector space as the skills data
    user_vector = vectorizer.transform([user_input])
    
    # Calculate cosine similarity between the user input and the courses' skills
    cosine_similarities = cosine_similarity(user_vector, skills_matrix)
    
    # Get the top 5 matching courses based on similarity scores
    top_indexes = cosine_similarities[0].argsort()[-5:][::-1]
    
    # Retrieve URLs, categories, and sub-categories for the top 5 courses
    recommended_courses = df.iloc[top_indexes][['Title','URL']]
    
    return recommended_courses

# Example: Get recommendations for a given skill input
user_input = "AI"
recommended_courses = recommend_courses(user_input)
print(recommended_courses)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle

# Function to load dataset from a file path
def load_dataset(file_path):
    """
    Function to load dataset from a CSV or other file format into a pandas DataFrame.
    Assumes the dataset has columns 'Skills', 'Title', and 'URL'.
    """
    df = pd.read_csv(file_path)

    # Replace NaN values in the 'Skills' column with an empty string
    df['Skills'] = df['Skills'].fillna('')

    return df

# Function to train and save the model
def train_and_save_model(df):
    """
    Function to train the TfidfVectorizer model on the 'Skills' column and save it to a pickle file.
    """
    vectorizer = TfidfVectorizer(stop_words='english')
    skills_matrix = vectorizer.fit_transform(df['Skills'])

    # Save the vectorizer to a pickle file for later use
    with open("tfidf_vectorizer.pkl", 'wb') as f:
        pickle.dump(vectorizer, f)

    return vectorizer, skills_matrix

# Function to get course recommendations based on user input
def recommend_courses(user_input, vectorizer, skills_matrix, df):
    """
    Function to predict course recommendations based on user input.
    Returns the top 5 courses based on cosine similarity.
    """
    # Transform the user input into the same vector space as the skills data
    user_vector = vectorizer.transform([user_input])

    # Calculate cosine similarity between the user input and the courses' skills
    cosine_similarities = cosine_similarity(user_vector, skills_matrix)

    # Get the top 5 matching courses based on similarity scores
    top_indexes = cosine_similarities[0].argsort()[-5:][::-1]

    # Retrieve Titles, URLs for the top 5 courses
    recommended_courses = df.iloc[top_indexes][['Title', 'URL']]

    return recommended_courses

# Main logic
if __name__ == "__main__":
    # Load the dataset from the provided file path
    file_path = r"C:\Users\pamru\Downloads\archive (4)\Online_Courses.csv"  # Replace with your actual file path
    df = load_dataset(file_path)

    # Check if model already exists; if not, train and save it
    try:
        with open("tfidf_vectorizer.pkl", 'rb') as f:
            vectorizer = pickle.load(f)
        skills_matrix = vectorizer.transform(df['Skills'])
        print("Successfully loaded pre-trained vectorizer.")
    except FileNotFoundError:
        print("No pre-trained vectorizer found, training a new one.")
        vectorizer, skills_matrix = train_and_save_model(df)

    # Example: Get recommendations for a given skill input
    user_input = "AI, deep learning, neural networks"
    recommended_courses = recommend_courses(user_input, vectorizer, skills_matrix, df)

    print("Recommended Courses:")
    print(recommended_courses)
```
